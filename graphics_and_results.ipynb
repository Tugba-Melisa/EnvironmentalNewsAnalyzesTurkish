```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('trainedilmis.csv')
# Assuming df is your DataFrame
cevrehaber = df[df['pseudo_label'] == 1]
uzakolan = df[df['pseudo_label'] != 0]

# Calculate the sizes for each category
sizes = [len(cevrehaber), len(uzakolan)]
labels = ['Çevre haberleri', 'Uzak Olan']
colors = ['#ff9999','#66b3ff']

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('çevre haberlerinin oranı ')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie chart
plt.show()
```

```python
from scipy.sparse import csc_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import IsolationForest
import pandas as pd
df=pd.read_csv('sentiment_data.csv')


# TF-IDF dönüşümü
vectorizer = TfidfVectorizer()
X_csc = vectorizer.fit_transform(df['cleaned_text'])

# CSC formatından CSR formatına dönüştürme
X = X_csc.toarray()

# Aykırı değer tespiti için Isolation Forest modelini oluşturma
model = IsolationForest(contamination=0.1)  # contamination aykırı değer oranını belirtir

# Modeli veri setine uygulama
model.fit(X)

# Aykırı değerleri tespit etme
df['Aykırı'] = model.predict(X)  # CSC formatında veriyi kullanıyoruz
df['Aykırı'] = df['Aykırı'].map({1: 0, -1: 1})  # 1: normal, -1: aykırı

print(df['Aykırı'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV dosyasını okuma
df = pd.read_csv('trainedilmis.csv')  # CSV dosyanızın adını buraya yazın

# Pseudo label değeri 1 olan haberlerin yüzdesini hesaplama
source_counts = df[df['pseudo_label'] == 1]['source'].value_counts()
total_source_counts = df['source'].value_counts()

# Yüzdelik hesaplama
source_percentages = (source_counts / total_source_counts) * 100

# İlk 15 kaynağı seçme
top_sources = source_percentages.nlargest(15)

# Grafik oluşturma
plt.figure(figsize=(12, 6))
top_sources.plot(kind='bar')
plt.title('Kaynaklara Göre Çevre Haberlerinin Yüzdesi')
plt.xlabel('Kaynak')
plt.ylabel('Haber Yüzdesi')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Tarih sütununu datetime formatına çevirme
df['date'] = pd.to_datetime(df['date'])

# Ay sütunu ekleme
df['month'] = df['date'].dt.to_period('M')

# Ay ay gruplama ve pseudo_label'a göre sayma
monthly_counts = df.groupby(['month', 'pseudo_label']).size().unstack(fill_value=0)

# Oransal hesaplama
monthly_percentages = (monthly_counts[1] / monthly_counts.sum(axis=1)) * 100

# Grafik oluşturma
plt.figure(figsize=(14, 8))
monthly_percentages.plot(kind='line', marker='o', color='b')
plt.title('Tarihsel Olarak Çevre Haberlerinin Oranı')
plt.xlabel('Ay')
plt.ylabel('Haber Yüzdesi (%)')
plt.grid(True)
plt.show()
```
